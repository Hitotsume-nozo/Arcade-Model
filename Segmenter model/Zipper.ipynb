```python
import shutil
import os

def zip_directory_simple(folder_path, output_filename):
    """
    Zips a directory using shutil.
    
    Args:
        folder_path (str): The path to the folder you want to zip.
        output_filename (str): The name of the resulting file (no .zip extension needed).
    """
    shutil.make_archive(output_filename, 'zip', folder_path)
    print(f"Created {output_filename}.zip successfully.")

zip_directory_simple('./final_evaluation', 'eval_metric_data')
```
